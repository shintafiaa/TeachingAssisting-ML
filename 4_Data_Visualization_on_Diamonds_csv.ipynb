Beri kode untuk menampilkan boxplot dari fitur-fitur numerik!

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Menampilkan boxplot dari setiap fitur numerik
numerical_features = df.select_dtypes(include=['float64', 'int64']).columns
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.set_palette("pink")
    sns.boxplot(x=df[feature])
    plt.title('Boxplot dari Fitur ' + feature)
    plt.show()
```

# Scatter Plot

ChatGPT prompt: beri kode untuk menampilkan scatter plot untuk semua fitur numerik

```python
# Mengatur gaya plot seaborn
sns.set(style="ticks")

# Membuat scatter plot untuk semua fitur numerik
sns.pairplot(df.drop(['cut', 'color', 'clarity'], axis=1), diag_kind='kde')
plt.suptitle("Scatter Plot untuk Fitur-Fitur Numerik", y=1.02, fontsize=15)
plt.show()
```

# Heat Map

ChatGPT prompt: beri kode untuk menampilkan heat map dari fitu-fitur numerik

```python
# Menghitung matriks korelasi antar fitur numerik
correlation_matrix = df.drop(['cut', 'color', 'clarity'], axis=1).corr()

# Membuat heat map dari matriks korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heat Map dari Korelasi antar Fitur Numerik', fontsize=15)
plt.show()
```
